```python
from ortools.linear_solver import pywraplp
ni=26
nj=15
nt=7
solver = pywraplp.Solver.CreateSolver('CP-SAT')
# solver.SetSolverSpecificParametersAsString('log_level=2')  # 2 代表详细日志级别
y = [[[solver.IntVar(0, 1, f'y_{t}_{i}_{j}') for j in range(nj)] for i in range(ni)] for t in range(nt)]
x = [[[solver.IntVar(0, 86, f'x_{t}_{i}_{j}') for j in range(nj)] for i in range(ni)] for t in range(nt)]

theta=0
# b = [solver.NumVar(0, solver.infinity(), f'b_{j}') for j in range(nj)]
B = [solver.NumVar(0, solver.infinity(), f'B_{t}') for t in range(nt)]
S = [solver.NumVar(0, solver.infinity(), f'S_{t}') for t in range(nt)]
for t in range(nt):
    B[t] = solver.Sum(bc.iloc[j] * x[t][i][j] for i in range(ni) for j in range(nj))
output =[[solver.NumVar(0, solver.infinity(), f'output_{t}_{j}') for j in range(nj)] for t in range(nt)]
remain =[[solver.NumVar(0, solver.infinity(), f'remain_{t}_{j}') for j in range(nj)] for t in range(nt)]
z1=[[solver.IntVar(0,1 , f'output_{t}_{j})') for j in range(nj)] for t in range(nt)]
z2=[[solver.IntVar(0,1 , f'output_{t}_{j})') for j in range(nj)] for t in range(nt)]

for t in range(nt):
    sum_output = solver.Sum([sc[j] * output[t][j] for j in range(nj)])
    sum_remain = theta*solver.Sum([sc[j] * remain[t][j] for j in range(nj)])
    for j in range(nj):
        # 约束2: output_{t}_{j} <= sum_{i=1} x_{ijt} * pro_{ij}
        sum_pro = solver.Sum(x[t][i][j] * pro.iloc[i,j] for i in range(ni))
        solver.Add(output[t][j] <= sum_pro)
        # b=sum_pro-Exp.iloc[j,0] 
        # M=1000000
        # solver.Add(remain[t][j] >= 0)
        # # 约束5: remain_{t}_{j} >= sum_{i=1} x_{ijt} * pro_{ij} - exp_{j}
        # solver.Add(remain[t][j] >= sum_pro - Exp.iloc[j,0])
        # solver.Add(b>=-M*(1-z1[t][j]))
        # solver.Add(remain[t][j]<=M*z1[t][j])
        solver.Add(output[t][j] <= Exp.iloc[j,0])
        solver.Add(remain[t][j]<=sum_pro+1000*z1[t][j])
        solver.Add(remain[t][j]<= 0+1000*z2[t][j])
        solver.Add(remain[t][j]>=sum_pro-Exp.iloc[j,0])
        solver.Add(remain[t][j]>=0)
        solver.Add(z1[t][j]+z2[t][j]==1) 
    S[t]=sum_output+sum_remain
# for i, idx in enumerate(x0.index):
#     for j, col in enumerate(x0.columns):
#         x[0][i][j] = x0.loc[idx, col]
# for i, idx in enumerate(x0.index):
#     for j, col in enumerate(x0.columns):
#         y[0][i][j] = y0.loc[idx, col]
for i in range(ni):
    for t in range(nt):
        constraint_expr = solver.Sum(x[t][i][j] for j in range(nj))
        solver.Add(constraint_expr <= area[i])
for t in range(nt): 
    if t==0:
        for i in range(ni):
            for j in range(nj):
                solver.Add(y[t][i][j] + y0.iloc[i,j] <= 1)
    else:
        for i in range(ni):
            for j in range(nj):
                solver.Add(y[t][i][j] + y[t -1][i][j] <= 1)
big_M = 10000
for t in range(nt):
    for i in range(ni):
        for j in range(nj):
            solver.Add(x[t][i][j] <= big_M * y[t][i][j])
for i in range(ni):
    for t in range(1, nt):
        if(t==1): 
            # 创建约束
            constraint_expr = solver.Sum(x[t][i][j] + x[t-1][i][j] + x0.iloc[i,j] for j in bean_class_indices)
            solver.Add(constraint_expr >= area[i])
        else:
            # 创建约束
            constraint_expr = solver.Sum(x[t][i][j] + x[t-1][i][j] + x[t-2][i][j] for j in bean_class_indices)
            solver.Add(constraint_expr >= area[i])
        
N = 3
for j in range(nj):
    for t in range(nt):
        # 创建约束
        constraint_expr = solver.Sum(y[t][i][j] for i in range(ni))
        solver.Add(constraint_expr <= N)
M = 0.3

for t in range(nt):
    for j in range(nj):
        for i in range(ni):
            # 创建约束
            constraint_expr = x[t][i][j] >= M * y[t][i][j] * area[i]
            solver.Add(constraint_expr)
solver.SetTimeLimit(120000)
solver.EnableOutput()
objective_expr = solver.Sum(S[t] for t in range(nt)) - solver.Sum(B[t] for t in range(nt))
solver.Maximize(objective_expr)
status = solver.Solve()
status
```

```python
# 输出结果
import numpy as np
if status == pywraplp.Solver.OPTIMAL:
    print('Optimal solution found.')
    print('Solution:')
    for t in range(nt):
        # print(f'S_{t} = {S[t].solution_value()}')
        # print(f'B_{t} = {B[t].solution_value()}')
        print(S[t].solution_value(),B[t].solution_value(),S[t].solution_value()-B[t].solution_value())
        x_np = np.array([[[x[t][i][j].solution_value() for j in range(nj)] for i in range(ni)] for t in range(nt)])
        y_np = np.array([[[y[t][i][j].solution_value() for j in range(nj)] for i in range(ni)] for t in range(nt)])
        # x_np= np.round(x_np)
elif status == pywraplp.Solver.INFEASIBLE:
    print('Problem is infeasible.')
    for t in range(nt):
        print(S[t].solution_value(),B[t].solution_value(),S[t].solution_value()-B[t].solution_value())
elif status == pywraplp.Solver.UNBOUNDED:
    print('Problem is unbounded.')
elif status == pywraplp.Solver.ABNORMAL:
    print('Solver encountered an abnormal termination.')
else:
    print('Solver status unknown.')
    print(status)
    for t in range(nt):
        print(S[t].solution_value(),B[t].solution_value(),S[t].solution_value()-B[t].solution_value())
        # print(f'B_{t} = {B[t].solution_value()}')
        # print(f'S_{t}-B_{t} = {S[t].solution_value()-B[t].solution_value()}')
    print('x values:')
    x_np = np.array([[[x[t][i][j].solution_value() for j in range(nj)] for i in range(ni)] for t in range(nt)])
    y_np = np.array([[[y[t][i][j].solution_value() for j in range(nj)] for i in range(ni)] for t in range(nt)])
    # x_df = pd.DataFrame(x_np[2], columns=[f'j_{j}' for j in range(nj)], index=[f'i_{i}' for i in range(ni)])
    # y_df = pd.DataFrame(y_np[1], columns=[f'j_{j}' for j in range(nj)], index=[f'i_{i}' for i in range(ni)])

file_path = r'D:\adocument\数学建模\24年国赛\CUMCM2024Problems\x10_5.xlsx'
with pd.ExcelWriter(file_path, engine='xlsxwriter') as writer:
    for i in range(nt):
        x_df = pd.DataFrame(x_np[i], columns=[f'j_{j}' for j in range(nj)], index=[f'i_{i}' for i in range(ni)])
        x_df.to_excel(writer, sheet_name=str(2023 + i))
```
