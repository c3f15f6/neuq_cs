```python
from ortools.linear_solver import pywraplp
ni=28
nj=41-15
nt=14
solver = pywraplp.Solver.CreateSolver('SCIP')
# solver.SetSolverSpecificParametersAsString('log_level=2')  # 2 代表详细日志级别
y = [[[solver.IntVar(0, 1, f'y_{t}_{i}_{j}') for j in range(nj)] for i in range(ni)] for t in range(nt)]
x = [[[solver.NumVar(0, 40, f'x_{t}_{i}_{j}') for j in range(nj)] for i in range(ni)] for t in range(nt)]

theta=0

S_T = [solver.NumVar(0, solver.infinity(), f'S_T[{T}]') for T in range(nt//2)]
B_T = [solver.NumVar(0, solver.infinity(), f'B_T[{T}]') for T in range(nt // 2)]
output = [[solver.NumVar(0, solver.infinity(), f'output[{T}]') for j in range(nj)] for T in range(nt//2 )]
remain = [[solver.NumVar(0, solver.infinity(), f'remain[{j}]') for j in range(nj)] for T in range(nt//2)]
for T in range(nt // 2):
    B_T[T] = solver.Sum(
        solver.Sum(bc.iloc[i,j] * x[T][i][j] for i in non_smart) +
        solver.Sum(bc.iloc[i,j]  * x[2*T] [i][j] for i in smart)
        for j in range(nj)
    )
for T in range(nt//2):
    sum_output = solver.Sum([sc[0][j] * output[T][j] for j in range(nj)])
    sum_remain = theta*solver.Sum([sc[0][j] * remain[T][j] for j in range(nj)])
    for j in range(nj):
        # 约束2: output_{t}_{j} <= sum_{i=1} x_{ijt} * pro_{ij}
        output_constraint_1 = solver.Sum(x[T][i][j] * pro1[0].iloc[i,j] for i in non_smart)
        output_constraint_2 = solver.Sum(x[2*T] [i][j]* pro1[0].iloc[i,j] for i in smart)
        solver.Add(output[T][j] <= output_constraint_1+output_constraint_2)
        # 约束3: output_{t}_{j} <= exp_{j}
        solver.Add(output[T][j] <= Exp[j])
        # 约束4: remain_{t}_{j} >= 0
        solver.Add(remain[T][j] >= 0)
        # 约束5: remain_{t}_{j} >= sum_{i=1} x_{ijt} * pro_{ij} - exp_{j}
        solver.Add(remain[T][j] >= output_constraint_1+output_constraint_2- Exp[j])
    S_T[T]=sum_output+sum_remain
output1 = [[solver.NumVar(0, solver.infinity(), f'output[{T}]') for j in range(nj)] for T in range(nt//2)]
remain1 = [[solver.NumVar(0, solver.infinity(), f'remain[{j}]') for j in range(nj)] for T in range(nt//2)]
B = [solver.NumVar(0, solver.infinity(), f'B_{t}') for t in range(nt//2)]
S = [solver.NumVar(0, solver.infinity(), f'S_{t}') for t in range(nt//2)]
for t in range(nt//2):  
    sum_output = solver.Sum([sc[1][j] * output1[t][j] for j in range(nj)])
    sum_remain = theta*solver.Sum([sc[1][j] * remain1[t][j] for j in range(nj)])
    for j in range(nj):
         # 约束2: output_{t}_{j} <= sum_{i=1} x_{ijt} * pro_{ij}
        output_constraint_2 = solver.Sum(x[2*t+1][i][j]* pro1[1].iloc[i,j] for i in smart)
        solver.Add(output1[t][j] <= output_constraint_2)
        # 约束3: output_{t}_{j} <= exp_{j}
        solver.Add(output1[t][j] <= Exp[j])
        # 约束4: remain_{t}_{j} >= 0
        solver.Add(remain1[t][j] >= 0)
        # 约束5: remain_{t}_{j} >= sum_{i=1} x_{ijt} * pro_{ij} - exp_{j}
        solver.Add(remain1[t][j] >= output_constraint_2- Exp[j])
    S[t] =sum_output+sum_remain 
for t in range(nt//2):
    B[t] = solver.Sum(bc.iloc[i,j] * x[2*t+1][i][j] for i in smart for j in range(nj))
# for i in range(ni):
#     for t in range(nt):
#         constraint_expr = solver.Sum(x[t][i][j] for j in range(nj))
#         solver.Add(constraint_expr <= area[i])

# for T in range(nt//2):  
#     for i in range(ni):
#         non_smart_sum=0 
#         smart_num=0
#         if i in smart:
#             non_smart_sum = solver.Sum(x[T][i][j] for j in range(nj))
#         else:
#             smart_sum = solver.Sum(x[2*T] [i][j]for j in range(nj))
#         # 添加约束
#         solver.Add(non_smart_sum + smart_sum <= area.iloc[i])
for T in range(nt//2):  
    for i in non_smart:
        non_smart_sum = solver.Sum(x[T][i][j] for j in range(nj))
        # 添加约束
        solver.Add(non_smart_sum <= area.iloc[i])
for t in range(nt):  # t = 3, 5, ..., 15
    for i in smart:
        smart_sum_t = solver.Sum(x[t][i][j] for j in range(nj))
        solver.Add(smart_sum_t <= area.iloc[i])
# for i in range(ni):
#     for t in range(2, nt):
#         # 创建约束
#         constraint_expr = solver.Sum(x[t][i][j] + x[t-1][i][j] + x[t-2][i][j] for j in bean_class_indices)
#         solver.Add(constraint_expr >= area[i])

# # 三年种豆
for t in range(4, nt): 
    if t==4: 
        for i in smart:
            bean_sum = solver.Sum(x[t][i][j] + x[t-1][i][j] + x[t-2][i][j] + x[t-3][i][j] + x[t-4][i][j] + x0.iloc[i,j]
                                  for j in bean_class_indices)
            solver.Add(bean_sum >= area.iloc[i])
    else:
        for i in smart:
            bean_sum = solver.Sum(x[t][i][j] + x[t-1][i][j] + x[t-2][i][j] + x[t-3][i][j] + x[t-4][i][j] + x[t-5][i][j] 
                                  for j in bean_class_indices)
            solver.Add(bean_sum >= area.iloc[i])
for T in range(1, nt//2):  
    if T==1: 
        for i in non_smart:
            bean_sum_non_smart = solver.Sum(x[T][i][j] + x[T-1][i][j] + x0.iloc[i,j] for j in bean_class_indices)
            solver.Add(bean_sum_non_smart >= area.iloc[i])
    else:
        for i in non_smart:
            bean_sum_non_smart = solver.Sum(x[T][i][j] + x[T-1][i][j] + x[T-2][i][j] for j in bean_class_indices)
            solver.Add(bean_sum_non_smart >= area.iloc[i])
# 相邻两季不同
for T in range(nt // 2 ): 
    if T==0:
        for i in non_smart:
            for j in [0]:
                solver.Add(y[T][i][j] + y0.iloc[i,j] <= 1)
    else:
        for i in non_smart:
            for j in [0]:
                solver.Add(y[T][i][j] + y[T-1][i][j] <= 1)
for t in range(nt): 
    if t==0: 
        for i in smart:
            for j in range(nj):
                solver.Add(y[t][i][j] + y0.iloc[i,j]<= 1)
    else:
        for i in smart:
            for j in range(nj):
                solver.Add(y[t][i][j] + y[t-1][i][j] <= 1)

# x,y的约束
big_M = 10000
for t in range(nt):
    for i in range(ni):
        for j in range(nj):
            solver.Add(x[t][i][j] <= big_M * y[t][i][j])
# 不能太分散
N=5
for T in range(nt//2):  # T = 1, 2, 3, ...
    for j in range(nj):
        non_smart_sum_y = solver.Sum(y[T][i][j] for i in non_smart)
        smart_sum_y = solver.Sum(y[2*T] [i][j]for i in smart)
        solver.Add(non_smart_sum_y + smart_sum_y <= N)
for t in range(1, nt, 2):  # t = 1, 3, 5, ...
    for j in range(nj):
        smart_sum_y_t = solver.Sum(y[t][i][j] for i in smart)
        solver.Add(smart_sum_y_t <= N)
M=0.4
for T in range(nt // 2):  # T = 1, 2, 3, ...
    for j in range(nj):
        for i in non_smart:
            solver.Add(x[T][i][j] >=area.iloc[i] *M * y[T][i][j])

# for t in range(nt):  # t = 1, 2, 3, ...
#     for j in range(nj):
#         for i in smart:
#             solver.Add(x[t][i][j]  >= area.iloc[i]*M* y[t][i][j])
solver.SetTimeLimit(300000)
solver.EnableOutput()
objective_expr =  solver.Sum(S_T[t]-B_T[t] for t in range(nt//2))+ solver.Sum(S[t]-B[t] for t in range(nt//2))

solver.Maximize(objective_expr)
status = solver.Solve()
status
```
