```python
import pandas as pd 
data=pd.read_excel(r"D:\adocument\数学建模\24年国赛\训练\2020年\index.xlsx")
fenlei_columns=['信誉评级','是否违约','公司类型'] 
for i in fenlei_columns: 
    data[i]=pd.factorize(data[i])[0]
data=data.drop(columns=['企业代号'])
fenlei_columns.remove('是否违约')
data
```

```python
x=data.drop(columns=['是否违约'])
y=data['是否违约']
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

columns_to_scale = x.columns.drop(fenlei_columns)
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
scaler=StandardScaler() 
x_train[columns_to_scale]=scaler.fit_transform(x_train[columns_to_scale]) 
x_test[columns_to_scale]=scaler.transform(x_test[columns_to_scale])
x_train
```

## 随机森林模型

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
from skopt import BayesSearchCV
param_space={
    'n_estimators':(10,100),
    'max_depth':(1,20), 
    'min_samples_split': (2,10), 
    'min_samples_leaf':(1,5)
}
rfs=RandomForestClassifier()
opt= BayesSearchCV(rfs,param_space, n_iter=32,random_state=42)
opt.fit(x_train,y_train)
```

```python
best_model=opt.best_estimator_
print("Best parameters found: ", opt.best_params_)
best_model.fit(x_train,y_train)
y_pred=best_model.predict(x_test)
# 评估模型
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### 粒子群优化

```python
from pyswarm import pso 
from sklearn.model_selection import cross_val_score
import numpy as np 
# 定义损失函数
def rf_cv_score(params): 
    n_estimators, max_depth, min_samples_split, min_samples_leaf= params 
    rf=RandomForestClassifier(
        n_estimators=int(n_estimators),
        max_depth=int(max_depth),
        min_samples_split=int(min_samples_split),
        min_samples_leaf=int(min_samples_leaf)
    )
    return -np.mean(cross_val_score(rf, x, y, cv=5, scoring='accuracy'))
lb=[50,10,5,1]
ub=[100, 20, 10, 5]
best_params, best_score= pso(rf_cv_score, lb, ub, swarmsize=20, maxiter=10) 
```

```python
print("Best parameters found: ", best_params,'\n','best score',best_score)
n_estimators, max_depth, min_samples_split, min_samples_leaf = best_params
best_rf = RandomForestClassifier(
    n_estimators=int(n_estimators),
    max_depth=int(max_depth),
    min_samples_split=int(min_samples_split),
    min_samples_leaf=int(min_samples_leaf)
)

# 在整个训练集上训练最终的模型
best_rf.fit(x, y)
y_pred=best_model.predict(x)
# 评估模型
print(confusion_matrix(y, y_pred))
print(classification_report(y, y_pred))
```

```python
import xgboost as xgb
from skopt.space import Real,Integer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, confusion_matrix
from skopt import BayesSearchCV
# 定义 XGBClassifier 模型
xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    # scale_pos_weight=1/3,
    eval_metric='logloss', 
    seed=42
)
param_space={
    'n_estimators': Integer(50,200),
    'max_depth': Integer(3,10), 
    'learning_rate': Real(0.01, 0.3, prior='uniform'),
    'subsample': Real(0.6, 1.0, prior='uniform'),
    'colsample_bytree': Real(0.6, 1.0, prior='uniform')
}
opt=BayesSearchCV(xgb_model,param_space, n_iter=32, scoring='roc_auc',cv=5,n_jobs=-1,random_state=42)
# 执行贝叶斯优化
opt.fit(x_train, y_train)
# 打印最佳参数和最佳分数
print("Best parameters found: ", opt.best_params_)
print("Best ROC AUC score: ", opt.best_score_)
```

```python
best_xgb_model=opt.best_estimator_
# 训练模型
best_xgb_model.fit(x_train, y_train)

# 预测
y_pred_prob = best_xgb_model.predict_proba(x_test)[:, 1]  # 获取预测的概率
threshold = 0.25
y_pred_class = (y_pred_prob  >= threshold).astype(int)# 设置概率投票
# 评估模型
print(confusion_matrix(y_test, y_pred_class))
accuracy = accuracy_score(y_test, y_pred_class)
print(f"Accuracy: {accuracy:.2f}")
# 打印分类报告
print(classification_report(y_test, y_pred_class))
# 计算并打印 AUC
auc = roc_auc_score(y_test, y_pred_prob)
print(f"AUC: {auc:.2f}")
y_pred_prob
```

## 逻辑回归-贝叶斯优化

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer
# 初始化逻辑回归模型
lgr = LogisticRegression(max_iter=1000, random_state=42)

# 定义超参数空间
param_space = {
    'C': Real(1e-6, 1e+6, prior='log-uniform'),  # 正则化强度
    'penalty': Categorical(['l1', 'l2']),        # 惩罚项类型
    'solver': Categorical(['liblinear', 'saga']) # 优化算法
}
# 设置贝叶斯优化
opt = BayesSearchCV(
    lgr,
    param_space,
    n_iter=32,  # 迭代次数
    scoring='accuracy',  # 评估指标
    cv=5,  # 交叉验证次数
    n_jobs=-1,  # 使用所有可用的CPU核
    random_state=42
)
# 拟合模型
opt.fit(x_train, y_train)
lr=opt.best_estimator_
```

```python
# 评估模型
y_pred = lr.predict(x)
print(confusion_matrix(y, y_pred))
print(classification_report(y, y_pred))
```
