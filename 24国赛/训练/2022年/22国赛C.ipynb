+ 风化前化学成分的预测

思路（以铅钡类为例）：
1. 分别计算风化和未风化的化学成分方差，我们认为**风化前后方差极小的化学成分是风化前后变化小的成分**（这里用小朱查的资料，理论上来说明风化前后会发生明显变化的化学成分更好）
2. 以风化前后变化小的数据为变量进行聚类分析，分别将风化前后样本分为k组
3. 找到组之间的对应关系，可以是对成分中心的数值的分析来分类划分等
4. 完成对应后我们认为组与组之间即是风化前后的关系（这里对组与组之间风化有差异的解释是：可能由于文物本身制造工艺、存在时长等因素影响）
5. 认为一组的风化前后变化程度相近，用均值计算变化程度，并应用到每一组的每个样本上预测风化前化学成分

```python
# 定义CLR变换函数
import numpy as np

def tran_CLR(data):
    # 计算数据的对数
    log_data=np.log(data)

    # 计算每个特征的均值
    mean_log_data = np.mean(log_data, axis=0)

    # 中心
    centered_log_data = log_data - mean_log_data

    # 计算每个特征的方差
    var_log_data = np.var(centered_log_data, axis=0)

    # 标准化
    normalized_log_data = centered_log_data / np.sqrt(var_log_data)
    return normalized_log_data
```

```python
import time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import nonzero, array
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA

def initialize_centroids(data, k):
    # 从数据集中随机选择k个点作为初始质心
    centers = data[np.random.choice(data.shape[0], k, replace=False)]
    return centers


def get_clusters(data, centroids):
    # 计算数据点与质心之间的距离，并将数据点分配给最近的质心
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    cluster_labels = np.argmin(distances, axis=1)
    return cluster_labels


def update_centroids(data, cluster_labels, k):
    # 计算每个簇的新质心，即簇内数据点的均值
    new_centroids = np.array([data[cluster_labels == i].mean(axis=0) for i in range(k)])
    return new_centroids


def k_means(data, k, T, epsilon):
    start = time.time()  # 开始时间，计时
    # 初始化质心
    centroids = initialize_centroids(data, k)
    t = 0
    while t <= T:
        # 分配簇
        cluster_labels = get_clusters(data, centroids)

        # 更新质心
        new_centroids = update_centroids(data, cluster_labels, k)

        # 检查收敛条件
        if np.linalg.norm(new_centroids - centroids) < epsilon:
            break
        centroids = new_centroids
        # print("第", t, "次迭代")
        t += 1
    print("用时：{0}".format(time.time() - start))
    return cluster_labels, centroids


# 计算聚类指标
def clustering_indicators(labels_true, labels_pred):
    if type(labels_true[0]) != int:
        labels_true = LabelEncoder().fit_transform(df[columns[len(columns) - 1]])  # 如果数据集的标签为文本类型，把文本标签转换为数字标签
    f_measure = f1_score(labels_true, labels_pred, average='macro')  # F值
    accuracy = accuracy_score(labels_true, labels_pred)  # ACC
    normalized_mutual_information = normalized_mutual_info_score(labels_true, labels_pred)  # NMI
    rand_index = rand_score(labels_true, labels_pred)  # RI
    ARI = adjusted_rand_score(labels_true, labels_pred)
    return f_measure, accuracy, normalized_mutual_information, rand_index, ARI


# 绘制聚类结果散点图
def draw_cluster(dataset, centers, labels):
    center_array = np.array(centers)
    attributes = len(dataset.columns)
    if attributes > 2:
        dataset = PCA(n_components=2).fit_transform(dataset)  # 如果属性数量大于2，降维
        center_array = PCA(n_components=2).fit_transform(center_array)  # 如果属性数量大于2，降维
    else:
        dataset = np.array(dataset)
    # 做散点图
    label = array(labels)
    plt.scatter(dataset[:, 0], dataset[:, 1], marker='o', c='black', s=7)  # 原图
    # plt.show()
    colors = np.array(
        ["#FF0000", "#0000FF", "#00FF00", "#FFFF00", "#00FFFF", "#FF00FF", "#800000", "#008000", "#000080", "#808000",
         "#800080", "#008080", "#444444", "#FFD700", "#008080"])
    # 循换打印k个簇，每个簇使用不同的颜色
    for i in range(k):
        plt.scatter(dataset[nonzero(label == i), 0], dataset[nonzero(label == i), 1], c=colors[i], s=7, marker='o')
    # plt.scatter(center_array[:, 0], center_array[:, 1], marker='x', color='m', s=30)  # 聚类中心
    plt.show()
```

```python
import random

# 设定扰动因子为x，则扰动后的化学成分数据在[(1-x)*含量，(1+x)*含量]范围内随机选取

ingredients = ['二氧化硅(SiO2)','氧化钠(Na2O)','氧化钾(K2O)','氧化钙(CaO)','氧化镁(MgO)','氧化铝(Al2O3)','氧化铁(Fe2O3)',
               '氧化铜(CuO)','氧化铅(PbO)','氧化钡(BaO)','五氧化二磷(P2O5)','氧化锶(SrO)','氧化锡(SnO2)','二氧化硫(SO2)']
ingredients_Pb = ['二氧化硅(SiO2)','氧化铅(PbO)','氧化钡(BaO)']

# data = tran_CLR(data_Pb[ingredients_Pb]) # 中心对数变换
data = data_Pb[ingredients_Pb]

for i in range(len(ingredients_Pb)):
    for x in np.arange(0, 1, 0.01):
        for j in range(len(data_Pb)): # 对数据进行扰动处理
            value = data.loc[j][i]
            data_new = data
            data_new.loc[j][i] = random.uniform((1-x)*value, (1+x)*value)
            # print(data.loc[j][i])
        # print(data)
        # 使用KMeans++算法进行聚类
        kmeans_plus_new = KMeans(n_clusters=3, init='k-means++', random_state=42) # 'k-means++' 是关键参数
        kmeans_plus_new.fit(np.array(data_new))

        labels = kmeans_plus.labels_
        labels_new = kmeans_plus_new.labels_

        if labels_new.any() != labels.any():
            break
        
    print(f"{ingredients_Pb[i]}的扰动范围是：",x)
```
